```python
import numpy as np 
import matplotlib.pyplot as plt
from vector_creation import *
from train import *
from test import *
```

## $\cos{\theta}$

```python
fulloutputCOS_train = train("data/cos_theta_train.txt", 1e-3, -0.99, 1.0, 20)
```

```python
d = fulloutputCOS_train[-1]
plt.stairs(abs(d))
plt.ylim(0,1.5)
plt.show()
```

```python
k = 1
s = fulloutputCOS_train[1]
tau = s[k]**2
```

```python
x_unfolded, x_true = test("data/cos_theta_test.txt", fulloutputCOS_train, t)
```

```python
first_bin = -0.99
last_bin = 1.0
bin_size = 20
bin_width = (last_bin - first_bin)/(bin_size)
bin_full = np.arange(first_bin, last_bin + bin_width, bin_width)

plt.stairs(x_true, bin_full, color='r', label='true')
plt.stairs(x_unfolded, bin_full, color='b', label='predicted')
plt.legend(loc='upper left')
plt.title("$\cos\\theta$ distribution")
plt.savefig("cos_distribution.png")
plt.show()
```

## $m_{tt}$

```python
mtt = np.loadtxt("data/mtt_train.txt")
min(mtt[:,0]),min(mtt[:,1]),max(mtt[:,0]),max(mtt[:,1])
```

**Note**
While choosing bins for the vectors, we should remember that measured distribution (b) can't have zero bin count, because the eigenvalues of the B covariance matrix (after SVDing) will be inverted to rescale. It doesn't matter whether true vector has any zero count.

```python
fulloutputMTT_train = train("data/mtt_train.txt", 1e-3, 400, 2000, 16)
```

```python
k = 1
s = fulloutputMTT_train[1]
tau = s[k]**2
```

```python
mtt_unfolded, mtt_true = test("data/mtt_test.txt", fulloutputMTT_train, tau)
```

```python
first_bin = 400
last_bin = 2000
bin_size = 16
bin_width = (last_bin - first_bin)/(bin_size)
bin_full = np.arange(first_bin, last_bin + bin_width, bin_width)

plt.stairs(mtt_true, bin_full, color='r', label='true')
plt.stairs(mtt_unfolded, bin_full, color='b', label='predicted')
# plt.stairs(x, bin_full, color='g', label='training x')
plt.legend(loc='upper right')
plt.title("$m_{t\\bar{t}}$ distribution")
plt.savefig("mtt_distribution.png")
plt.show()
```
